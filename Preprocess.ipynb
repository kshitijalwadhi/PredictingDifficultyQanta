```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("Modified_2.csv")
```

```python
df.head()
```

```python
df = df[['last_line','difficulty']]
```

```python
df.head()
```

```python
df
```

```python
df['difficulty'].value_counts()
```

```python
qs_list = []
difficulty_list = []

for idx,row in df.iterrows():
    if row['difficulty'] == 'College' or ('college' in row['difficulty']):
        qs_list.append(row['last_line'])
        difficulty_list.append('college')
    elif row['difficulty'] == 'HS' or ('high_school' in row['difficulty']):
        qs_list.append(row['last_line'])
        difficulty_list.append('high_school')
```

```python
qs_list[0]
```

```python
qs_dict = {'qs':qs_list,'difficulty':difficulty_list}
```

```python
new_df = pd.DataFrame(qs_dict)
```

```python
new_df.head()
```

```python
new_df.to_csv("final.csv",index=False)
```
